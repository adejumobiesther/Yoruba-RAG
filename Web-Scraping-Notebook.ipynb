```python
url = 'https://www.theyorubablog.com/'
req = Request(url=url,headers={'user-agent': 'my-app/0.0.1'})
response = urlopen(req)
html = bs(response)
```

```python
mydiv = html.find("div", {"class": "menu-menu-1-container"}).find("ul")
```

```python
mydiv
```

```python
# Find all the links to different articles on the blog
posts = []
n=0
article_links = mydiv.find_all('a', href=True)
for article in article_links[3:]:
    article_url = article['href']
    article_req = Request(url=article_url,headers={'user-agent': 'my-app/0.0.1'})
    article_response = urlopen(article_req)
    article_html = bs(article_response)
    # Find the <div> element with id='content'
    content_div = article_html.find('div', id='content')
    for article in content_div.find_all('article'):
        article_content = article.find('div', class_='entry-content') 
        if article_content == None:
            continue
        posts.append(article_content.get_text())
    n+=1
    print(f'Finished {n} link')
```

```python
posts
```

```python
# Remove "\xa0" from each string in the list
cleaned_post = [s.replace('\xa0', '').strip() for s in posts]
```
